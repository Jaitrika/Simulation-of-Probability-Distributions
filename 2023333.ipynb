```python
import matplotlib.pyplot as plt
import numpy as np

def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.5, numRepeatsOfExp=1, numTrialsInAnExp=1, profitLoss=[1, 0]):
    observation = randomNumGen.random((numRepeatsOfExp, numTrialsInAnExp))
    index_ge1 = np.less_equal(observation, bernoulliParam_p)
    observation[index_ge1] = profitLoss[0]
    observation[np.invert(index_ge1)] = profitLoss[1]
    return observation

v_numTrialsInAnExp = 20
v_numRepeatsOfExp = 1
v_bernoulliParam_p = 0.2

# Plot outcomes against trial number
plt.figure(figsize=(8, 5))
observation = aBernoulliExperiment(np.random.default_rng(), bernoulliParam_p=v_bernoulliParam_p, numRepeatsOfExp=v_numRepeatsOfExp, numTrialsInAnExp=v_numTrialsInAnExp)
plt.step(range(1, v_numTrialsInAnExp + 1), observation.ravel(), where='mid')

# Customize plot
plt.xlabel('Trial Number')
plt.ylabel('Outcome (1 for Heads, 0 for Tails)')
plt.title('Coin Toss Outcomes (20 Trials)')
plt.xticks(range(1, v_numTrialsInAnExp + 1))  # Specify integer ticks on x-axis
plt.grid(True)

# Show plot
plt.show()
```

this plot shows outputs (0 for tails , 1 for heads) when a coin is tossed 20 times whose probablity of showing heads is 0.2

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing seed for reproducibility
np.random.seed(42) 

# Function to simulate a bernoulli experiment
def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.5, numRepeatsOfExp=1, numTrialsInAnExp=1, profitLoss=[1, 0]):
    observation = randomNumGen.random((numRepeatsOfExp, numTrialsInAnExp))
    index_ge1 = np.less_equal(observation, bernoulliParam_p)
    observation[index_ge1] = profitLoss[0]
    observation[np.invert(index_ge1)] = profitLoss[1]
    return observation

# Function to simulate a binomial experiment
def aBinomialExperiment(randomNumGen, binomialParam_n=5, binomialParam_p=0.5, numberOfBinomialExperiments=1, profitLoss_Bern=[1,0]):
    outcomesOfBernTrails = aBernoulliExperiment(randomNumGen, bernoulliParam_p=binomialParam_p, numRepeatsOfExp=numberOfBinomialExperiments, numTrialsInAnExp=binomialParam_n, profitLoss=profitLoss_Bern)
    outcomeOfBinomialExp = np.sum(outcomesOfBernTrails, axis=1)
    return outcomeOfBinomialExp

# Function to simulate a Poisson experiment
def aPoissonExperiment(randomNumGen, lam=1, numberOfExperiments=1):
    outcomesOfPoissonExperiments = np.random.poisson(lam=lam, size=(1, numberOfExperiments))
    return outcomesOfPoissonExperiments

rng = np.random.default_rng()

numberOfBinomialExps = 10000
numberOfPoissonExps = 10000

binomial_outcomes = []
poisson_outcomes = []

alphas = rng.uniform(low=0, high=100, size=4)

for alpha in alphas:
    p = 0.2  # Fixing  p 
    n = int(alpha / p)  

    binomial_outcomes.append(aBinomialExperiment(rng, binomialParam_n=n, binomialParam_p=p, numberOfBinomialExperiments=numberOfBinomialExps))

    poisson_outcomes.append(aPoissonExperiment(rng, lam=alpha, numberOfExperiments=numberOfPoissonExps))

# Plot histograms
fig, axs = plt.subplots(4, 2, figsize=(12, 16))

for i in range(4):
    # Plot histogram for binomial experiment
    axs[i, 0].hist(binomial_outcomes[i], bins=np.arange(0, np.max(binomial_outcomes[i]) + 2) - 0.5, alpha=0.7,  edgecolor='black', color='blue', density=True, width=0.8)
    axs[i, 0].set_title(f'Binomial (n={int(alphas[i] / 0.2)}, p=0.2)')  # Dynamically calculating n

    # Plot histogram for Poisson experiment
    axs[i, 1].hist(poisson_outcomes[i][0], bins=np.arange(0, np.max(poisson_outcomes[i]) + 2) - 0.5, alpha=0.7, edgecolor='black', color='blue', density=True, width=0.8)
    axs[i, 1].set_title(f'Poisson (alpha={alphas[i]})')

for ax in axs.flat:
    ax.set_xlabel('Random Variable')
    ax.set_ylabel('Probability')

plt.tight_layout()
plt.show()
```

here histograms of binomial are compared shape wise to resemble that of the poisson distribution for each of the four randomly generated alphas.
we observe as alpha increases the poisson and binomial distributions look similar, symmetric and also bell- shaped.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate a Poisson experiment
def aPoissonExperiment(lam=1, numberOfExperiments=1):
    outcomesOfPoissonExperiments = np.random.poisson(lam=lam, size=(1, numberOfExperiments))
    return outcomesOfPoissonExperiments

alphas = np.random.uniform(1, 20, size=2)
sequence_lengths = [5, 50, 500, 5000]

for alpha in alphas:
    plt.figure(figsize=(12, 8))
    plt.suptitle(f'Alpha = {alpha}')
    for i, length in enumerate(sequence_lengths, start=1):
        poisson_rvs = []
        for _ in range(1000):
            experiments = aPoissonExperiment(lam=alpha, numberOfExperiments=length)
            poisson_rvs.append(experiments)
            
        poisson_rvs = np.concatenate(poisson_rvs, axis=0)
        sums = np.sum(poisson_rvs, axis=1)
        plt.subplot(2, 2, i)
        plt.hist(sums, bins=20, density=True, alpha=0.7, edgecolor='black')
        plt.title(f'Sequence Length = {length}')
        plt.xlabel('Sum of Poisson RVs')
        plt.ylabel('Probability Density')
    plt.tight_layout()
    plt.show()

```


this displays convergence to a bell-shaped curve and clustering around the mean as chosen the sequence lengths are (5, 50, 500, 5000). 
As the sequence length increases, the distribution of the sum of Poisson random variables tends to resemble a normal distribution hence CLT is verified . 
Additionally, the histogram shows that the frequencies cluster more tightly around the mean as the sequence length increases hence LLN is also verified.

we can verify this as we know that mean is here the sum of alphas and from MGF we can say that the new aplha is the sum of rest of the alphas.

These observations verify both the law of large numbers and the central limit theorem for the sum of iid Poisson random variables.
